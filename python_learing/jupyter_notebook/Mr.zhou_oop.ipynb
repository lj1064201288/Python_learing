### 定义一个学生类,有下面的类属性
- 姓名
- 年龄
- 成绩(语文,数学,英语)每科成绩类型为整数
- 类方法:
    - 获取学生的姓名:get_name()返回类型:str
    - 获取学生的年纪:get_age()返回类型:int
    - 返回3门科目中最高的分数,get_course()返回类型:int

```python
class Student(object):
    def __init__(self, name, age, socres):
        self.name = name
        self.age = age
        self.socres = socres
        
    def get_name(self):
        return self.name
    
    def get_age(self):
        return self.age
    
    def get_course(self):
        return max(self.socres)

ss = Student("liujun", 18, (87, 95, 100))
print(ss.get_name())
print(ss.get_age())
print(ss.get_course())
```

### 定义一个字典类型:DictClass, 完成如下功能
##### 1.删除某个key del_key(key)
##### 2.判断某个键是否在字典里, 如果再返回键对应的值, 不在则返回not found get_dict()
##### 3.返回键组成的列表,返回类型: list get_key()
##### 4.合并字典,并且返回合并字典的values组成的列表,返回类型 list, update_dict()

```python
class DictClass(object):
    def __init__(self, dict):
        self.dict = dict
        
    def del_key(self, key):
        # 判断key是否在字典当中
        if key in self.dict.keys():
            self.dict.pop(key) 
            print(key + '已删除...')
        else:
            print(key + '不在字典当中...')
            
    def get_dict(self, key):
        if key not in self.dict.keys():
            return 'Not Found ' + key
        else:
            return self.dict[key]
        
    def get_list(self):
        return self.dict.keys()
        
    def update_dict(self, dict2):
        self.dict = dict(self.dict, **dict2)
        return self.dict.values()
        
            
books = {
    'python': '$56',
    'C':'$48',
    'Java':'$78',
    'PHP':'$45'
}

book = DictClass(books)
book.get_dict('go')
keys = book.get_list()
print(keys)
books2 = {
    'go':'$67'
}

book.update_dict(books2)

```

### 定义一个列表的操作类 Listinfo
## 包括的方法
##### 1.列表元素添加:add_key(),添加的内容必须是数字或者字符串
##### 2.列表元素取值:get_key()
##### 3.列表合并:update_list(list)
##### 4.删除并且返回最后一个元素:del_key()

```python
class ListInfo(object):
    def __init__(self, list_val):
        self.list = list_val
        
    def add_key(self,key):
        # 判断的数值必须是字符串或者数字
        if isinstance(key, (str, int)):
            self.list.append(key)
            print(self.list)
            return 'OK'
        else:
            print('请输入字符串或者数字')
        
    def get_key(self, index):
        # 判断传入的索引是否超出了列表
        if index >= 0 and index < len(self.list):
            return self.list[index]
        elif index > len(self.list) -1:
            print('超出索引列表')
    
    def update_list(self, list2):
        self.list.extend(list2)
        return self.list
    
    def del_key(self):
        # 首先判断列表中是否偶元素
        if len(self.list) >= 0:
            return self.list.pop(-1)
        else:
            return '列表中已经没有元素了!'
    
        
ls = [1,2,3,4,5]
ls2 = [6,7,8,9]

list1 = ListInfo(ls)
list1.add_key(6)
```

### 定义一个集合的操作类
- 包括的方法:
    - 集合元素添加:add_setinfo()
    - 集合的交集: get_intersection()
    - 集合的并集: get_ubion()
    - 集合的差集: del_difference()
    

```python
class SetInfo(object):
    def __init__(self, my_set):
        self.myset = my_set
        
    def add_setinfo(self, key):
        self.myset.add(key)
        return self.myset
    
    def get_intersetction(self, unioninfo):
        if isinstance(unioninfo, set):
            return self.myset & unioninfo
        else:
            print('请传入集合类型的数据')
        
    def get_union(self, unioninfo):
        if isinstance(unioninfo, set):
            return self.myset | unioninfo
        else:
            print('请传入集合类型的数据')
    
    def difference(self, unioninfo):
        if isinstance(unioninfo, set):
            return self.myset - unioninfo
        else:
            print('请传入集合类型的数据')
my_set = {1,2,3,4,5}
my_set1 = {2,3,5,6,7}

myset = SetInfo(my_set)
myset.get_intersetction(my_set1)
# myset.get_union(my_set1)
# myset.difference(my_set1)
```

- 学员小李在查看了自己在Python3的成绩列表然后退出了

```python
# 校区列表
Course_list = []

# 创建校区主类
class School(object):
    def __init__(self, school_name):
        '''
        school_name:学校的名称
        '''
        self.school_name = school_name
        # 学生列表
        self.students_list = []
        # 老师列表
        self.teachers_list = []
        global Course_list
    
    def hire(self, obj):
        '''
        obj: 一个老师的对象
        '''
        # 添加一个老师到老师的列表当中
        self.teachers_list.append(obj.name)
        print('我们现在聘请了一个新老师{}'.format(obj.name))
    
    def enroll(self, obj):
        '''
        obj:一个学生的对象
        '''
        # 添加一个学生到学生的列表当中
        self.students_list.append(obj.name)
        print('我们又有了一个新学员{}'.format(obj.name))

# 年级,课程的类
class Grade(School):
    def __init__(self, school_name, grade_code, grade_course):
        '''
        school_name:学校名称
        grade_code:学习的期数
        grade_course: 学习的课程
        '''
        # 继承父类的属性
        super(Grade,self).__init__(school_name)
        self.code = grade_code
        self.course = grade_course
        self.member = []
        # 课程的列表
        Course_list.append(self.course)
        
        print('我们现在有了{}学院的{}期的{}课程'.format(self.school_name, self.code, self.course))
        
    def course_info(self):
        print('课程大纲{} 是day01, day02, day03'.format(self.course))

# 创建一个北京校区Python的课程
Python = Grade('北京', 3, 'python')
# 创建一个成都校区的Linux的课程
linux = Grade('成都', 1,'linux')

# 添加学校成员
class School_member(object):
    def __init__(self, name, age, sex, role):
        '''
        name: 姓名
        age: 年龄
        sex: 性别
        role: 角色
        '''
        self.name = name
        self.age = age
        self.sex = sex
        self.role = role
        self.course_list = []
        print('我叫{}, 我是一个{}'.format(self.name, self.role))
    
# 初始化学生的id
stu_num_id = 00

# 添加学生的类
class Students(School_member):
    def __init__(self, name, age, sex, role, course):
        
        # 集成主类的属性
        super(Students, self).__init__(name, age, sex, role)
        # 使id学号变成全局变量
        global stu_num_id
        stu_num_id += 1
        
        # 学号组成: 学校名字 + S + 课程期数 + 学号id(对数字进行填充,当只有一位数的时候给前面填充0)
        stu_id = course.school_name + 'S' + str(course.code) + str(stu_num_id).zfill(2)
        # zifill 填充的作用, 当只有一位数的时候前面填充0, 只能对str类型做操作
        
        # 学员的id
        self.id = stu_id
        self.mark_list = []
        
    def study(self, course):
        '''
        course: 课程名称
        '''
        print('我来这里学习{}课, 我的学号是{}'.format(course.course, self.id))
        
    def pay(self, course):
        '''
        给学习的课程缴费
        course: 课程名称
        '''
        print('我教了1000块钱给{}'.format(course.course))
        
        # 将缴费的课程添加到学员的报名课程里面
        self.course_list.append(course.course)
        
    def Praise(self, obj):
        '''
        点评
        obj:接受一个学员的对象
        '''
        print('{}觉得真棒{}课真棒'.format(self.name, obj.name))
        
    def mark_check(self):
        # 查看自己所报名的课程
        for i in self.mark_list.items():
            print(i)
            
    def out(self):
        # 退出
        print('我离开了... ')

# 老师的id
tea_num_id = 00

class Teachers(School_member):
    def __init__(self, name, age, sex, role, course):
        # 继承学校成员的主类属性
        super(Teachers, self).__init__(name, age, sex, role)
        # 将id变成全局变量
        global tea_num_id
        tea_num_id += 1
        tea_id = course.school_name + 'T' + str(course.code) + str(tea_num_id).zfill(2)

        self.id = tea_id
        
    # 老师讲的课程
    def teach(self, course):
        '''
        course: 课程名称
        '''
        print('我来这里讲{}这门课, 我的id是{}'.format(course.course, self.id))


    def record_mark(self, Date, course, obj, level):
        '''
        上课记录
        Date:第几期课程
        course:课程名称
        obj: 一个学员的对象
        level: 评分顶级
        '''
        obj.mark_list['Day' + Date] = level
            
a = Students('小张', 18, 'M', 'student', Python)
Python.enroll(a)
a.study(Python)
a.pay(Python)

b = Students('小王', 22, 'F', 'student', Python)
Python.enroll(b)
b.study(Python)
b.pay(Python)

c = Teachers('小周', 30, 'M', 'teacher', Python)
Python.hire(c)
c.teach(Python)
c.record_mark(2, 'python', Python, 'A')
```

```python
import random
import tkinter

class RandomBall():
    def _init_(self,canvas,scrnwidth,scrnheight):
        self.canvas=canvas
        self.xpos=random.randint(10, int(scrnwidth)-20)
        self.ypos=random.randint(10, int(scrnheight)-20)
        self.xvelocity=random.randint(2,5)
        self.yvelocity=random.randint(2,5)
        self.radius=random.randint(20,120)
        c=lambda: random.randint(0,255)
        self.color='#%02x%02x%02x'%(c(),c(),c())
        self.scrnwidth=scrnwidth
        self.scrnheight=scrnheight
    def create_ball(self):
        x1=self.xpos-self.radius
        y1=self.ypos+self.radius
        x2=self.xpos+self.radius
        y2=self.ypos-self.radius
        self.item=self.canvas.create_oval(x1,y1,x2,y2,fill=self.color,outline=self.color)
    def move_ball(self):
        self.xpos+=self.xvelocity
        self.ypos+=self.yvelocity
        if  self.xpos+self.radius>=self.scrnwidth\
                or self.xpos+self.radius<=20:
            self.xvelocity=-self.xvelocity
        if  self.ypos+self.radius>=self.scrnheight\
                or self.ypos+self.radius<=20:
            self.yvelocity=-self.yvelocity
            
        self.canvas.move(self.item,self.xvelocity,self.yvelocity)
            
class ScreenSave():
    def _init_(self):
        self.balls=[]
        self.num_balls=random.randint(4,20)
        self.root=tkinter.TK()
        self.root.overrideredirect(1)
        self.root.attributes('-alpha',0.3)
        self.root.bind=('<motion',self.myquit)
        self.root.bind=('<key>',self.myquit)
        w,h=self.root.winfo_screenwidth(),self.winfo_screenheight()
        self.canvas=tkinter.canvas(self.root,width=w,height=h)
        self.canvas.pack()
        for i in range(self.num_balls):
            ball=RandomBall(self.canvas,scrnwidth=w,scrnheight=h)
            ball.create_ball()
            self.balls.append(ball)
        self.run_screen_saver()
        self.root.mainloop()
        
    def run_screen_saver(self):
        for ball in self.balls:
            ball.move_ball()
        self.canvas.after(20,self.run_screen_saver)
    def myquit(self,e):
        self.root.destroy()
if __name__==" __main __":
    ScreenSaver()
```
