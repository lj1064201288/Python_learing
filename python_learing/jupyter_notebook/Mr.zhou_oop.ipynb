### 定义一个学生类,有下面的类属性
- 姓名
- 年龄
- 成绩(语文,数学,英语)每科成绩类型为整数
- 类方法:
    - 获取学生的姓名:get_name()返回类型:str
    - 获取学生的年纪:get_age()返回类型:int
    - 返回3门科目中最高的分数,get_course()返回类型:int

```python
class Student(object):
    def __init__(self, name, age, socres):
        self.name = name
        self.age = age
        self.socres = socres
        
    def get_name(self):
        return self.name
    
    def get_age(self):
        return self.age
    
    def get_course(self):
        return max(self.socres)

ss = Student("liujun", 18, (87, 95, 100))
print(ss.get_name())
print(ss.get_age())
print(ss.get_course())
```

### 定义一个字典类型:DictClass, 完成如下功能
##### 1.删除某个key del_key(key)
##### 2.判断某个键是否在字典里, 如果再返回键对应的值, 不在则返回not found get_dict()
##### 3.返回键组成的列表,返回类型: list get_key()
##### 4.合并字典,并且返回合并字典的values组成的列表,返回类型 list, update_dict()

```python
class DictClass(object):
    def __init__(self, dict):
        self.dict = dict
        
    def del_key(self, key):
        # 判断key是否在字典当中
        if key in self.dict.keys():
            self.dict.pop(key) 
            print(key + '已删除...')
        else:
            print(key + '不在字典当中...')
            
    def get_dict(self, key):
        if key not in self.dict.keys():
            return 'Not Found ' + key
        else:
            return self.dict[key]
        
    def get_list(self):
        return self.dict.keys()
        
    def update_dict(self, dict2):
        self.dict = dict(self.dict, **dict2)
        return self.dict.values()
        
            
books = {
    'python': '$56',
    'C':'$48',
    'Java':'$78',
    'PHP':'$45'
}

book = DictClass(books)
book.get_dict('go')
keys = book.get_list()
print(keys)
books2 = {
    'go':'$67'
}

book.update_dict(books2)

```

### 定义一个集合的操作类

    - 集合的交集: get_intersection()
    - 集合的并集: get_ubion()
    - 集合的差集: del_difference()
    

```python
class SetInfo(object):
    def __init__(self, my_set):
        self.myset = my_set
        
    def add_setinfo(self, key):
        self.myset.add(key)
        return self.myset
    
    def get_intersetction(self, unioninfo):
        if isinstance(unioninfo, set):
            return self.myset & unioninfo
        else:
            print('请传入集合类型的数据')
        
    def get_union(self, unioninfo):
        if isinstance(unioninfo, set):
            return self.myset | unioninfo
        else:
            print('请传入集合类型的数据')
    
    def difference(self, unioninfo):
        if isinstance(unioninfo, set):
            return self.myset - unioninfo
        else:
            print('请传入集合类型的数据')
my_set = {1,2,3,4,5}
my_set1 = {2,3,5,6,7}

myset = SetInfo(my_set)
myset.get_intersetction(my_set1)
# myset.get_union(my_set1)
# myset.difference(my_set1)
```

### 定义一个门票系统
- 门票的原价是100元
- 当周末的时候门票涨20%
- 小孩子半价
- 计算2个成年人和一个小孩子平日的票价

```python
class Ticket():
    def __init__(self, weekend=False, child=False):
        self.exp = 100
        if weekend:
            self.inc = 1.2
        else:
            self.inc = 1
            
        if child:
            self.discount = 0.5
        else:
            self. discount = 1
            
    def cal_price(self, num):
        return self.exp * self.inc * self.discount * num
    
adult = Ticket()
child = Ticket(child=True)

print('两个成年人和一个小孩子平日的票价为{}'.format(adult.cal_price(2) + child.cal_price(1)))
```

```python
class Door():
    def __init__(self, weekend=False, child=False):
        self.price = 100
        if weekend:
            self.inc = 1.2
        else:
            self.inc = 1
        if child:
            self.dis = 0.5
        else:
            self.dis = 1
            
    def cal_price(self, num):
        '''
        num:人数的个数
        '''
        return self.price * self.inc * self.dis * num
    
au = Door()
child = Door(child=True)

print("两个成人加上一个小孩平常的价格是{}".format(au.cal_price(2) + child.cal_price(1)))
```

### 龟龟游戏
- 假设游戏场景为范围(x,y)为0<=x<=10.0<=y<=10
- 游戏生成1只乌龟和10条鱼
- 他们的移动方向均随机
- 乌龟的最大移动能力是2(乌龟可以随机选择移动1还是2), 鱼的最大移动能力为1
- 当移动到场景边缘,自动想反方向移动
- 乌龟初始化体力为100(上限)
- 乌龟每一栋一次,体力消耗1
- 当乌龟和鱼重叠,乌龟吃掉鱼, 乌龟体力增加20
- 鱼不计算体力
- 当乌龟体力值为0或者鱼的数量为0时, 游戏结束

```python
import random 

class Turtle(object):
    def __init__(self):
        self.power = 100
        
        # 初始化乌龟的位置
        self.x = random.randint(0, 10)
        self.y = random.randint(0,10)
        
    def move(self):
        new_x = random.choice([1,2,-1,-2]) + self.x
        new_y = random.choice([1,2,-1,-2]) + self.y
        
        # 判断乌龟的位置是否超出了边界
        
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else:
            self.x = new_x
            
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
            
        self.power -= 1
        return (self.x, self.y)
    
    def eat(self):
        self.power += 20
        if self.power >= 100:
            self.power = 100
            
class Fish():
    def __init__(self):
        self.x = random.randint(0, 10)
        self.y = random.randint(0, 10)
        
    def move(self):
        new_x = random.choice([1,-1]) + self.x
        new_y = random.choice([1,-1]) + self.y
        
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else:
            self.x = new_x
            
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
            
        return (self.x, self.y)
            
turtle = Turtle()
fish = []

for num in range(10):
    new_fish = Fish()
    fish.append(new_fish)
    
while True:
    if not len(fish):
        print('鱼儿被吃完了!!!')
        break
    if not turtle.power:
        print('乌龟没有体力了!!')
        break
        
    pos = turtle.move()
    
    # 在迭代器中做列表的删除元素是非常危险的,经常会出现一些意想不到的问题, 因为迭代器是直接引用列表元素的数据的操作
    # 所以 我们这里把列表拷贝一份传给迭代器,然后在对原列表做操作
    for new_fish in fish[:]:
        if new_fish.move() == pos:
            turtle.eat()
            fish.remove(new_fish)
            print("有一条鱼被吃掉了!")
            
        
```

```python
import random

class Wugui(object):
    def __init__(self):
        self.x = random.randint(0, 10)
        self.y = random.randint(0, 10)
        
        self.power = 100
        
    def move(self):
        new_x = random.choice([-1,-2,1,2]) + self.x
        new_y = random.choice([-1,-2,1,2]) + self.y
        
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else:
            self.x = new_x
            
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
            
        self.power -= 1
            
        return (self.x, self.y)
            
    def eat(self):
        self.power += 20
        if self.power > 100:
            self.power = 100
            
class Yu():
    def __init__(self):
        self.x = random.randint(0, 10)
        self.y = random.randint(0, 10)
        
    def move(self):
        new_x = random.choice([-1,-2,1,2]) + self.x
        new_y = random.choice([-1,-2,1,2]) + self.y
        
        if new_x < 0:
            self.x = 0 - (new_x - 0)
        elif new_x > 10:
            self.x = 10 - (new_x - 10)
        else:
            self.x = new_x
            
        if new_y < 0:
            self.y = 0 - (new_y - 0)
        elif new_y > 10:
            self.y = 10 - (new_y - 10)
        else:
            self.y = new_y
            
        return (self.x, self.y)
            
turtle = Wugui()
fishs = []

for num in range(10):
    fish = Yu()
    fishs.append(fish)
    
while True:
    if not len(fishs):
        print('鱼儿被吃完了,游戏结束!')
        break
    if not turtle.power:
        print('乌龟没有体力了,游戏结束!!')
        break
    
    pos = turtle.move()
    
    for new_fish in fishs[:]:
            if pos == new_fish.move():
                turtle.eat()
                fishs.remove(new_fish)
                print('有一条鱼儿被吃了...')
```

#### 定义一个点(point)和直线(Line)类, 使用getLine方法获取两点过程直线的长度

```python
import math

class Point():
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        
    def get_x(self):
        return self.x
    def get_y(self):
        return self.y
    
class Line():
    def __init__(self, p1, p2):
        self.x = p1.get_x() - p2.get_x()
        self.y = p1.get_y() - p2.get_y()
        
        self.lening = math.sqrt(self.x**2 + self.y**2)
        
    def getLine(self):
        return self.lening
    
p1 = Point(2,3)
p2 = Point(5,7)

line = Line(p1, p2)
line.getLine()
```
