- URL = 'http://www.cbooo.cn/'

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.cbooo.cn/year?year=2018"
datas = requests.get(url)
```

```python
datas = requests.get(url).text
```

```python
# 利用bs4对数据进行解析
soup = BeautifulSoup(datas, 'lxml')
```

```python
# 获取数据集合
movies_table = soup.find_all('table', {'id':'tbContent'})[0]
print(movies_table)
```

```python
movies = movies_table.findAll('tr')
movies
```

```python
# 获取电影名称
names = [tr.find_all('td')[0].a.get('title') for tr in movies[1:]]
names
```

```python
# 获取电影url地址
hrefs = [tr.find_all('td')[0].a.get('href') for tr in movies[1:]]
hrefs
```

```python
# 获取电影类型
types = [tr.find_all('td')[1].string for tr in movies[1:]]
types
```

```python
# 获取总票房数据
boxoffice = [int(tr.find_all('td')[2].string) for tr in movies[1:]]
boxoffice
```

```python
# 获取平均票价
mean_price = [int(tr.find_all('td')[3].string) for tr in movies[1:]]
mean_price
```

```python
# 获取场均人次
mean_people = [int(tr.find_all('td')[4].string) for tr in movies[1:]]
mean_people
```

```python
# 获取国家及地区
contries = [tr.find_all('td')[5].string for tr in movies[1:]]
contries
```

```python
# 获取上映日期
times = [tr.find_all('td')[6].get_text() for tr in movies[1:]]
times
```

```python
# 获取导演
def getInfo(url):
    datas = requests.get(url)
    soup = BeautifulSoup(datas.text, 'lxml')
    # 获取导演
    daoyan = soup.select('dl.dltext dd')[0].get_text()
    return daoyan
```

```python
directors = [getInfo(url).strip() for url in hrefs]
directors
```

```python
df = pd.DataFrame({
    'name':names,
    'href':hrefs,
    'type':types,
    'boxoffice(万)':boxoffice,
    'mean_price':mean_price,
    'mean_pople':mean_people,
    'contrie': contries,
    'time':times,
    'directir':directors
})
```

```python
df.head(5)
```

```python
# 数据存储
df.to_csv('../datas/tables/movies.csv')
```
